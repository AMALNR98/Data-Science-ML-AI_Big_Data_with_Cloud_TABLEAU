```python
import numpy as np
import pandas as pd
customer = pd.read_csv('/content/customer1.txt')
customer.columns = ['id', 'first_name', 'last_name', 'age', 'profession', 'location']
customer
```

```python
location_count = customer.groupby('location')['location'].count().sort_values(ascending=False)
location_count
```

```python
import matplotlib.pyplot as plt
plt.hist(location_count)
```
