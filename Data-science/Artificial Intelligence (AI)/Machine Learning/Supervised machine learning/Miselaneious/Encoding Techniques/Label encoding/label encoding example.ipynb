**encoding**

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
credit_score['Gender'] = le.fit_transform(credit_score['Gender'])
credit_score['Education'] = le.fit_transform(credit_score['Education'])
credit_score['Marital Status'] = le.fit_transform(credit_score['Marital Status'])
credit_score['Home Ownership'] = le.fit_transform(credit_score['Home Ownership'])
credit_score['Credit Score'] = le.fit_transform(credit_score['Credit Score'])
```

```python
credit_score
```

```python
x = credit_score.iloc[:,:-1].values
y = credit_score.iloc[:,-1].values
```

```python
x
```

```python
x.ndim
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.30, random_state = 42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(x_train)
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
```

```python
from sklearn.svm import SVC
model = SVC()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
score = accuracy_score(y_test, y_pred)
score
```
