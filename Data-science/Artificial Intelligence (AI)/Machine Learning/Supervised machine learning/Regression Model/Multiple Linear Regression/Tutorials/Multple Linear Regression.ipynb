Multiple number of indipendent variables and one dependent variable

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
advertising = pd.read_csv('/content/advertising.csv')
advertising
```

```python
advertising.head()
```

```python
x = advertising.iloc[:,:-1].values
y = advertising.iloc[:,-1].values
x
```

plot graph : we use seaborn
- in seaborn we can plot scaterplot and prediction line at a time

```python
#TV vs Sales
import seaborn as sns
sns.regplot(x = advertising['TV'], y = advertising['Sales'], color = 'red')
#or
#sns.regplot(x = advertising['TV'], y = y)
```

```python
# Radio vs Sales
sns.regplot(x = advertising['Radio'], y = advertising['Sales'], color = 'blue')
```

```python
# News paper vs Sales
sns.regplot(x = advertising['Newspaper'], y = advertising['Sales'], color = 'yellow')
```

Training And Testing Data

```python
from pandas.core.common import random_state
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.30, random_state = 42)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.linear_model import LinearRegression
# Create Object(model)
model = LinearRegression()
# fit model
model.fit(x_train, y_train)
# Predict value
y_pred = model.predict(x_test)
y_pred
```

prediction

```python
print(model.predict([[150.1, 45.5, 75.4]]))
```

Data Frame

```python
df = pd.DataFrame({'Actual_value':y_test, 'Predicted_value':y_pred, 'Difference': y_test - y_pred})
df
```

Print slope

```python
print('slope is', model.coef_) #TV. Radio, NewsPaper
```

```python
# To print seperatly
print('slop of TV is :', model.coef_[0])
print('slop of Radio is :', model.coef_[1])
print('slop of NewsPaper is :', model.coef_[2])
```

Or using zip

```python
print("Slope is")
list(zip(x, model.coef_))
```

Pring constant

```python
print("constant is ", model.intercept_)
```

Perfomance Evaluation

```python
# Mean Absoulte Percentage error
from sklearn.metrics import mean_absolute_percentage_error
print("Error percentage is ", mean_absolute_percentage_error(y_test, y_pred))
```

```python
# R2_Score
from sklearn.metrics import r2_score
print("R2 score is", r2_score(y_test, y_pred))
```
