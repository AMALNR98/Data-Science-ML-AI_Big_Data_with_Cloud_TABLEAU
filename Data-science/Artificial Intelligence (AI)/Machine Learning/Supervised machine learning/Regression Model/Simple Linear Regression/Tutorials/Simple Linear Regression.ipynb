Simple Linear Regression is one of the fundamental techniques in regression analysis, used to model the ``relationship between a single independent variable and a continuous dependent variable.``

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

salary_data = pd.read_csv('/content/Salary_Data.csv')
salary_data
```

seperate x and y
when we using regression model we don't have to create x and y as list, so avoid 'values()'

```python
x = salary_data.iloc[:,:-1].values
y = salary_data.iloc[:,-1].values
x
```

plot x and y

```python
plt.scatter(x,y)
plt.xlabel('Year of exp')
plt.ylabel("Salary")
plt.title('Exp vs salary')
```

testing and training data

```python
from pandas.core.common import random_state
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.30, random_state = 42)
```

- In regression model there is no need for scaling.
- So don't do normalization

***Model Creation***

```python
from sklearn.linear_model import LinearRegression
# Create Object(model)
model = LinearRegression()
# fit model
model.fit(x_train, y_train)
# Predict value
y_pred = model.predict(x_test)
y_pred
```

print actual value

```python
y_test
```

Prediction

```python
print(model.predict([[3.9]]))
```

***Plot Regression line***

```python
import matplotlib.pyplot as plt
plt.scatter(x_train, y_train, color = 'red')
plt.plot(x_test, y_pred) # Best fit line
plt.xlabel("x_train")
plt.ylabel("y_train")
```

Here we draw the best fit line based on
- y = mx+c
 - where
   - y = y_pred
   - x_test
   - m = slop or coefficient
   - c = constant

```python
print('slope or coefficient is', model.coef_)  # coef_
print("constant is ", model.intercept_)
```

How to evaluate performace measurements
---
- In Regression model there is no confusion matirx
- Error or bias
  1. MAE(Mean Absolute Error)
  2. MAPE(Mean Absolute Percentage Error)
  3. MSE(Mean Squared Error)
  4. RMSE (Root Mean Squared Error)
  5. R2_score :


***Mean Absolute Error***

```python
from sklearn.metrics import mean_absolute_error
print("MAE is", mean_absolute_error(y_test, y_pred))
```

***Mean Absolute Percentage Error***

```python
from sklearn.metrics import mean_absolute_percentage_error
print("MAPE is", mean_absolute_percentage_error(y_test, y_pred))
```

The error may upto 30%

***Mean Squared Error***

```python
from sklearn.metrics import mean_squared_error
value = mean_squared_error(y_test, y_pred)
print(value)
```

***Root Mean Squared Error***

There is no specific function for RMSE
find square root by numpy

```python
print(np.sqrt(value))
```

***R2_Score***

```python
from sklearn.metrics import r2_score
print("R2 score is", r2_score(y_test, y_pred))
```

Give importance to Mean Absoulte Error and R2_Score
