```python
import numpy as np
import pandas as pd
iris_flower_data = pd.read_csv('/content/Iris.csv')
iris_flower_data
```

Here we don't need Id for training, so drop id coulumn

```python
iris = iris_flower_data.drop(['Id'], axis = 1)
iris
```

Set input to x and output to y, and set it as an array(for that add '.values') output always should be 1 dimensional

```python
x = iris.iloc[:, :-1].values
y = iris.iloc[:, -1].values
x.ndim
```

```python
y.ndim
```

Set training and testin data

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.30)
```

```python
x_train
```

```python
y_train
```

```python
x_test
```

```python
y_test
```

Normalization

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(x_train)
normalized_x_train_data = scaler.transform(x_train)
normalized_x_test_data = scaler.transform(x_test)
```

```python
normalized_x_train_data
```

```python
normalized_x_test_data
```

for create model
  - create object
  - assign value for k as 7
  - To create model fit training data
  - here u_predic is to check if the machine is trained or not

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=7)
knn.fit(normalized_x_train_data, y_train)
y_pred = knn.predict(normalized_x_test_data)
y_pred
```

```python
y_test
```

Creating confusion matrix

```python
from sklearn.metrics import confusion_matrix, accuracy_score
matrix_con = confusion_matrix(y_test, y_pred)
matrix_con
```

Here we have 3 output options so the confusion matrix will be 3*3 matrix

Calcuate accuracy score

```python
score = accuracy_score(y_test, y_pred)
score
```

Prediction

```python
print(knn.predict([[4.7, 4.6, 1.4, 0.3]]))
```

To plot confusion matrix

```python
from sklearn.metrics import ConfusionMatrixDisplay
lables =  ['Iris-virginica', 'Iris-versicolor', 'Iris-setosa']
# The order must be in y_test
cmd = ConfusionMatrixDisplay(matrix_con, display_labels = lables)
cmd.plot()
```
