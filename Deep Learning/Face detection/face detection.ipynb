```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('/home/amal/Documents/Data-Science-ML-AI_Big_Data_with_Cloud_TABLEAU/Deep Learning/Face detection/data/1.jpeg')
```

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img_rgb)
```

```python
cv2.imshow('image',img)
cv2.waitKey(5000)
cv2.destroyAllWindows()
```

```python
img_gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
face_casecade = cv2.CascadeClassifier('/home/amal/Documents/Data-Science-ML-AI_Big_Data_with_Cloud_TABLEAU/Deep Learning/Face detection/data/haarcascade_frontalface_default.xml')
```

```python
faces = face_casecade.detectMultiScale(img_gray)
```

```python
faces
```

```python
for (x, y, w, h) in faces:
    cv2.rectangle(img,(x,y),(w+h,y+h), color = (0,255,0), thickness=3)
```

```python
cv2.imshow('face detected',img)
cv2.waitKey(5000)
cv2.destroyAllWindows()
```
