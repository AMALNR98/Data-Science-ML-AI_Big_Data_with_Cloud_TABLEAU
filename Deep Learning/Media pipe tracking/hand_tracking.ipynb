```python
import cv2
import matplotlib.pyplot as plt
import mediapipe as mp
```

```python
img = cv2.imread('/home/amal/Documents/Data-Science-ML-AI_Big_Data_with_Cloud_TABLEAU/Deep Learning/Media pipe tracking/Data/4.jpeg')
```

```python
cv2.imshow('image',img)
cv2.waitKey(5000)
cv2.destroyAllWindows()
```

```python
img_converted = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img_converted)
```

Hand tracking

```python
mp_hands = mp.solutions.hands
hands = mp_hands.Hands(static_image_mode= True, max_num_hands= 2,min_detection_confidence=0.3)
mp_drawing = mp.solutions.drawing_utils
```

```python
results = hands.process(img_converted)
```

```python
print(results.multi_handedness)
```

```python
if results.multi_hand_landmarks:    # checking if multi hand landmarks ouucred
    for hand_no, hand_landmarks in enumerate(results.multi_hand_landmarks): #Because of 2 hands
        mp_drawing.draw_landmarks(
            image=img_converted,
            landmark_list=hand_landmarks,
            connections=mp_hands.HAND_CONNECTIONS
            )
    plt.imshow(img_converted)    
```
