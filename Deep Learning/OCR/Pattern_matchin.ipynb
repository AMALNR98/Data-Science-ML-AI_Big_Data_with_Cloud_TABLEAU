```python
import cv2
import pytesseract
from pytesseract import Output
```

```python
img = cv2.imread('/home/amal/Documents/Data-Science-ML-AI_Big_Data_with_Cloud_TABLEAU/Deep Learning/OCR/Data/Invoice.jpeg')
```

```python
cv2.imshow('Invoice', img)
cv2.waitKey(5000)
cv2.destroyAllWindows()
```

```python
data = pytesseract.image_to_data(img, output_type=Output.DICT)
```

```python
n_boxes = len(data['text'])
n_boxes
```

Drawing rectangles over teaded texts

```python
img = cv2.imread('/home/amal/Documents/Data-Science-ML-AI_Big_Data_with_Cloud_TABLEAU/Deep Learning/OCR/Data/Invoice.jpeg')
for i in range(n_boxes):
    if data['conf'][i]>= 70:
        x, y, w, h = data['left'][i], data['top'][i], data['width'][i],data['height'][i]
        cv2.rectangle(img,(x,y),(x+y, y+h),(0,233,0),3)
```

```python
cv2.imshow('readed text', img)
cv2.waitKey(5000)
cv2.destroyAllWindows()
```

```python
import re
```

```python
date_pattern = '^(0[1-9]|[12][0-9]|3[01])/(0[1-9]|1[012])/(19|20)\d\d$'
mail_pattern =	'^[a-z0-9](\.?[a-z0-9]){5,}@example\.net$'
```

collecting date and mail

```python
img = cv2.imread('/home/amal/Documents/Data-Science-ML-AI_Big_Data_with_Cloud_TABLEAU/Deep Learning/OCR/Data/Invoice.jpeg')
for i in range(n_boxes):
    if data['conf'][i]>= 70:
        if re.match(date_pattern,data['text'][i]):
            x, y, w, h = data['left'][i], data['top'][i], data['width'][i],data['height'][i]
            cv2.rectangle(img,(x,y),(x+y, y+h),(0,233,0),3)
cv2.imshow('readed text', img)
cv2.waitKey(5000)
cv2.destroyAllWindows()
```

```python
img = cv2.imread('/home/amal/Documents/Data-Science-ML-AI_Big_Data_with_Cloud_TABLEAU/Deep Learning/OCR/Data/Invoice.jpeg')
for i in range(n_boxes):
    if data['conf'][i]>= 70:
        if re.match(mail_pattern,data['text'][i]):
            x, y, w, h = data['left'][i], data['top'][i], data['width'][i],data['height'][i]
            cv2.rectangle(img,(x,y),(x+y, y+h),(0,233,0),3)
cv2.imshow('readed text', img)
cv2.waitKey(5000)
cv2.destroyAllWindows()
```
