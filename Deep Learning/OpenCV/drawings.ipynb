```python
img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)  
plt.imshow(img_rgb)
```

Drawing rectangle

```python
rect_img = cv2.rectangle(img_rgb, pt1=(105,100),pt2=(10,150),color=(255,0,0),thickness=5)
plt.imshow(rect_img)
```

```python
img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)  
rect_img = cv2.rectangle(img_rgb, pt1=(105,100),pt2=(10,50),color=(0,255,0),thickness=1)
plt.imshow(rect_img)
```

To fill rectangle

```python
img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)  
rect_img = cv2.rectangle(img_rgb, pt1=(105,100),pt2=(10,50),color=(255,61,180),thickness=-1)
rect_img = cv2.rectangle(img_rgb, pt1=(60,150),pt2=(10,50),color=(20,161,80),thickness=-1)
rect_img = cv2.rectangle(img_rgb, pt1=(50,110),pt2=(10,50),color=(255,161,270),thickness=-1)

plt.imshow(rect_img)
```

Circle

```python
img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)  
cir_img = cv2.circle(img_rgb, center=(230,60), radius=60, thickness=5, color=(254,108,45))
plt.imshow(cir_img)
```
